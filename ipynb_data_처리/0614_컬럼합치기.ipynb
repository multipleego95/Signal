```python
import csv
import pandas as pd

# Path to the CSV file
csv_f = '원본_0613_cluster6_vectors.csv'

df = pd.read_csv(csv_f)

df
```

```python
# 모든 컬럼 값을 하나의 컬럼으로 합치기
# 컬럽합치기 = '' 안에는 구분자 넣기  
컬럼합치기 = df.apply(lambda x: ','.join(map(str, x)), axis=1)

# 새로운 데이터프레임 생성
vectors_0614 = pd.DataFrame({'vectors': 컬럼합치기})

vectors_0614
```

```python
vectors_0614.to_csv('클러스터6_벡터합친거.csv')
```
