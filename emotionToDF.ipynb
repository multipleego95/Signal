```python
import pandas as pd

# CSV 파일 경로
csv_file_path1 = "results_ss.csv"
csv_file_path2 = "results_sslast1.csv"
csv_file_path3 = "results_sslast2.csv"
csv_file_path4 = "results_sslast3.csv"
csv_file_path5 = "results_sslast4.csv"
csv_file_path6 = "results_sslast5.csv"
csv_file_path7 = "results_sslast6.csv"
csv_file_path8 = "results_sslast7.csv"
csv_file_path9 = "results_sslast8.csv"

# CSV 파일 불러오기
df1 = pd.read_csv(csv_file_path1)
df2 = pd.read_csv(csv_file_path2)
df3 = pd.read_csv(csv_file_path3)
df4 = pd.read_csv(csv_file_path4)
df5 = pd.read_csv(csv_file_path5)
df6 = pd.read_csv(csv_file_path6)
df7 = pd.read_csv(csv_file_path7)
df8 = pd.read_csv(csv_file_path8)
df9 = pd.read_csv(csv_file_path9)

# 데이터 합치기
df_list = [df1, df2, df3, df4, df5, df6, df7, df8, df9]
df_all = pd.concat(df_list, ignore_index=True)
```

```python
import pandas as pd

# 감정 데이터 리스트
emotions = ['기쁨', '당황', '분노', '불안', '상처', '슬픔']

# 그룹화 및 감정 데이터 개수 계산
grouped = df_all.groupby(['novel_no', '예측 감정']).size().unstack(fill_value=0)

# 새로운 데이터프레임 생성
new_df = pd.DataFrame({'novel_no': grouped.index})

# 감정 데이터 개수 열 추가
for i, emotion in enumerate(emotions):
    new_df[emotion] = grouped.iloc[:, i]

print(new_df)
```
