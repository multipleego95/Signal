```python
import csv
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd

def calculate_similarity(data):
    # 데이터를 데이터 프레임으로 변환
    data = pd.DataFrame(data).iloc[1:,1:]

    # TF-IDF 벡터화
    # vectorizer = TfidfVectorizer()
    # tfidf_matrix = vectorizer.fit_transform(data)

    # 코사인 유사도 계산
    # similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
    similarity_matrix = cosine_similarity(data, data)

    return similarity_matrix

# CSV 파일 읽기
data = []
with open('원본_0613_cluster6_vectors.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        data.append(row)

# 행별 유사도 계산(1번 군집 기준 : 1604 * 1604)
similarity_matrix = calculate_similarity(data)

# 유사한 인덱스 추출
n_similar_indices = 6  # 유사한 인덱스 개수
similar_indices = []
for i, row in enumerate(similarity_matrix):
    top_indices = np.argsort(-row)[0:n_similar_indices+1]  # 가장 유사한 인덱스 추출 (자기 자신은 제외)
    similar_indices.append(top_indices)

# 결과 출력
for i, indices in enumerate(similar_indices):
    print(f"인덱스 {i}과 가장 유사한 {n_similar_indices}개 인덱스:")
    for index in indices:
        print(index)
```

```python
import csv
import numpy as np

# 결과를 CSV 파일로 저장
result_file = '테스트_인덱스포함_클러스터6_similar_indices.csv'
with open(result_file, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Index'] + list(range(n_similar_indices + 1)))  # 헤더 작성
    for i, indices in enumerate(similar_indices):
        writer.writerow([i] + indices.tolist())  # 인덱스와 유사한 인덱스들 저장

print(f"유사한 인덱스를 {result_file} 파일로 저장했습니다.")
```
