```python
import csv
from konlpy.tag import Kkma

def extract_nouns(sentence):
    kkma = Kkma()
    nouns = kkma.nouns(sentence)
    return [noun for noun in nouns if len(noun) > 1]  # 한 글자인 명사는 제외
print("1")

files = {
    '전처리_Cluster_no1_0613.csv': '클러스터no1',
    '전처리_Cluster_no2_0613.csv': '클러스터no2',
    '전처리_Cluster_no3_0613.csv': '클러스터no3',
    '전처리_Cluster_no4_0613.csv': '클러스터no4',
    '전처리_Cluster_no5_0613.csv': '클러스터no5',
    '전처리_Cluster_no6_0613.csv': '클러스터no6'
}
print("2")
cluster_results = {}  # 클러스터 결과를 담을 딕셔너리

for file, variable in files.items():
    with open(file, 'r', encoding='utf-8') as f:
        reader = csv.reader(f)
        next(reader)  # 첫 번째 행은 컬럼 이름이므로 건너뜁니다.
        nouns_list = []
        for row in reader:
            second_column = row[1]  # 두 번째 컬럼 데이터 추출
            nouns = extract_nouns(second_column)  # 형태소 분석 및 명사 추출
            nouns_list.extend(nouns)  # 추출한 명사를 리스트에 추가

        noun_counts = {}
        for noun in nouns_list:
            noun_counts[noun] = noun_counts.get(noun, 0) + 1

        # 명사 빈도를 기준으로 순위를 매깁니다.
        sorted_nouns = sorted(noun_counts.items(), key=lambda x: x[1], reverse=True)

        # 결과를 딕셔너리에 저장
        result = [(rank + 1, noun, count) for rank, (noun, count) in enumerate(sorted_nouns[100:201])]
        cluster_results[variable + '_result'] = result
print("3")


# 결과 출력 및 변수에 저장
clusterno1_result = cluster_results['클러스터no1_result']
clusterno2_result = cluster_results['클러스터no2_result']
clusterno3_result = cluster_results['클러스터no3_result']
clusterno4_result = cluster_results['클러스터no4_result']
clusterno5_result = cluster_results['클러스터no5_result']
clusterno6_result = cluster_results['클러스터no6_result']
print("4")

# 결과 출력
print("클러스터no1 결과:")
for rank, noun, count in clusterno1_result:
    print(f"{rank}. {noun}: {count}")
print()
print("5")
```
